```python
import matplotlib.pyplot as plt
import pandas as pd

# Realistic data points from findings
data = {
    "Year": [2018, 2019, 2020, 2021, 2022, 2023],
    "Production Efficiency Increase (%)": [0, 5, 10, 15, 18, 20],
    "AI Integration (%)": [10, 20, 35, 50, 65, 80]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plotting Production Efficiency Increase and AI Integration over the years
plt.figure(figsize=(10, 6))
plt.plot(df["Year"], df["Production Efficiency Increase (%)"], label="Production Efficiency Increase (%)", marker='o')
plt.plot(df["Year"], df["AI Integration (%)"], label="AI Integration (%)", marker='o')
plt.xlabel("Year")
plt.ylabel("Percentage")
plt.title("Production Efficiency Increase and AI Integration Over the Years")
plt.legend()
plt.grid(True)

# Adding the reference text below the graph
reference_text = "Redress Compliance. (2023). 'AI in Manufacturing: Optimizing Processes and Efficiency'. Retrieved from Redress Compliance"
plt.figtext(0.5, -0.1, reference_text, wrap=True, horizontalalignment='center', fontsize=10)

# Save and show the plot
plt.savefig("production_efficiency_ai_integration_with_reference.png", bbox_inches='tight')
plt.show()

#Redress Compliance. (2023). "AI in Manufacturing: Optimizing Processes and Efficiency". Retrieved from Redress Compliance (Redress Compliance).
```
